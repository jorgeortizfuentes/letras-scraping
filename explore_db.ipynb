```python
from models import Artists, Songs, engine
from sqlalchemy.orm import sessionmaker
import pandas as pd

session = sessionmaker(bind=engine)()

# Get all songs 
songs = session.query(Songs).all()

# Get all song with lyrics not None 
songs = session.query(Songs).filter(Songs.lyrics != None).all()
# Convert the songs to a list of dictionaries 
songs = [song.__dict__ for song in songs]

df = pd.DataFrame(songs)
df["title"]
```

```python
# Convert all lyrics to None 
songs = session.query(Songs).all()
for s in songs:
    s.lyrics = None
session.commit()
```
